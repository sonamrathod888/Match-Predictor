```python
lines = []
with open(r'Bat.csv') as file:
    reader = csv.reader(file,delimiter=',')
    next(reader)
    for row in reader:
        row = list(map(lambda x: x.rstrip('*') if '*' in x else x,row))
        row = list(map(lambda x: '0.0' if '-' in x else x,row))
        row = list(map(lambda x:x if ' ' in x else float(x),row))
        #print(row)
        lines.append(row)
    print(lines)
        #print(len(row))
        #print(row)
```

```python
lines2 = []
for i in (lines):
    lines2.append([i[6],i[8]])
print(lines2)
```

```python
vecAssembler = VectorAssembler(inputCols=["avg","sr"], outputCol="features")
```

```python
final = []
for i in enumerate(lines2):
    final.append([i[1][0],i[1][1]])
print(final)

df1 = spark.createDataFrame((lines2),["avg", "sr"])
```

```python
batsman_clusters = []
dataa = sc.parallelize(final)
model = KMeans.train(dataa, 8, maxIterations=10,initializationMode="random")
print(model)
for i in range(len(final)):
    print(model.predict(np.array(final[i])))
    batsman_clusters.append([lines[i][0],model.predict(np.array(final[i]))])
```
