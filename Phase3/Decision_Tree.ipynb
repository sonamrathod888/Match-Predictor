```python

from __future__ import print_function
from pyspark import SparkContext
# $example on$
from pyspark.mllib.tree import DecisionTree, DecisionTreeModel
from pyspark.mllib.regression import LabeledPoint
from numpy import array

sc.stop()
if __name__ == "__main__":

    sc = SparkContext(appName="p3dt")
    sc.setLogLevel("ERROR")
        
    def clp1(f):
        bowler_cluster=int(f[0])
        bowler_economy=float(f[1])
        bowler_average=float(f[2])
        batsmen_1_cluster=int(f[3])
        batsmen_1_average=float(f[4])
        batsmen_1_strikerate=float(f[5])
        batsmen_2_cluster=int(f[6])
        batsmen_2_average=float(f[7])
        batsmen_2_strikerate=float(f[8])
        run=int(f[9])
        wicket=int(f[10])
        return LabeledPoint(run,[bowler_cluster,bowler_economy,bowler_average,batsmen_1_cluster,batsmen_1_average,batsmen_1_strikerate,batsmen_2_cluster,batsmen_2_average,batsmen_2_strikerate])
    def clp2(f):
        bowler_cluster=int(f[0])
        bowler_economy=float(f[1])
        bowler_average=float(f[2])
        batsmen_1_cluster=int(f[3])
        batsmen_1_average=float(f[4])
        batsmen_1_strikerate=float(f[5])
        batsmen_2_cluster=int(f[6])
        batsmen_2_average=float(f[7])
        batsmen_2_strikerate=float(f[8])
        run=int(f[9])
        wicket=int(f[10])
        return LabeledPoint(wicket,[bowler_cluster,bowler_economy,bowler_average,batsmen_1_cluster,batsmen_1_average,batsmen_1_strikerate,batsmen_2_cluster,batsmen_2_average,batsmen_2_strikerate])
        
            
    data = sc.textFile("output.csv")
    csvdata=data.map(lambda x:x.split(","))
    td1=csvdata.map(clp1)
    td2=csvdata.map(clp2)
    m1 = DecisionTree.trainRegressor(td1, categoricalFeaturesInfo={0:10,3:10,6:10},impurity='variance',maxDepth=14, maxBins=30)
    m2 = DecisionTree.trainRegressor(td2, categoricalFeaturesInfo={0:10,3:10,6:10},impurity='variance',maxDepth=14, maxBins=30)
    m1.save(sc, "runs")
    m2.save(sc, "wickets")
    print(m1.toDebugString())
    sc.stop()
```
